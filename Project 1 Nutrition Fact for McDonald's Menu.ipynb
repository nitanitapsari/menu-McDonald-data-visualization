# Soal No 3
1. Bagaimana rata-rata nutrisi (vitamin A, vitamin C, Kalsium dan Besi) antara menu Crispy Chicken dan Grilled Chicken?
2. Bagaimana rata-rata kalori pada menu Crispy Chicken dan Grilled Chicken?

```python
CrispyChicken = df[df['Item'].str.contains('Crispy Chicken')] #mengambil kolom item dan baris yg ada 'Crispy Chicken'
CrispyChicken_Cal = pd.DataFrame({'Item': CrispyChicken.Item, 'Calories': CrispyChicken.Calories}) #mengambil item crispy dan calorinya

#nutrisi
CrispyChicken_Nut = pd.DataFrame({'Item': CrispyChicken.Item, 'Vit A': CrispyChicken['Vitamin A (% Daily Value)'], 'Vit C': CrispyChicken['Vitamin C (% Daily Value)'], 'Calcium': CrispyChicken['Calcium (% Daily Value)'], 'Iron': CrispyChicken['Iron (% Daily Value)']})
```

```python
avg_CrispyChicken_Cal = CrispyChicken.Calories.mean(axis=0) #menghitung averange
print ("Kalori di Crispy Chicken Rata-rata: ",avg_CrispyChicken_Cal)
```

```python
avg_CrispyChicken_vita = round(CrispyChicken_Nut['Vit A'].mean(axis=0), 2)
avg_CrispyChicken_vitc = round(CrispyChicken_Nut['Vit C'].mean(axis=0), 2)
avg_CrispyChicken_calc = round(CrispyChicken_Nut['Calcium'].mean(axis=0), 2)
avg_CrispyChicken_iron = round(CrispyChicken_Nut['Iron'].mean(axis=0), 2)
print("Rata-rata VIT A, VIT C, CALCIUM, IRON di CRISPY  CHICKEN: ", 
      avg_CrispyChicken_vita, avg_CrispyChicken_vitc, avg_CrispyChicken_calc, avg_CrispyChicken_iron)
```

```python
GrilledChicken = df[df['Item'].str.contains('Grilled Chicken')] #mengambil kolom item dan baris yg ada 'Grilled Chicken'
GrilledChicken_Cal = pd.DataFrame({'Item': GrilledChicken.Item, 'Calories': GrilledChicken.Calories}) #mengambil item grilled dan calorinya

GrilledChicken_Nut = pd.DataFrame({'Item': GrilledChicken.Item, 'Vit A': GrilledChicken['Vitamin A (% Daily Value)'], 'Vit C': GrilledChicken['Vitamin C (% Daily Value)'], 'Calcium': GrilledChicken['Calcium (% Daily Value)'], 'Iron': GrilledChicken['Iron (% Daily Value)']})

#rata2 kalori
avg_GrilledChicken_Cal = GrilledChicken.Calories.mean(axis=0)

#rata2 Nutrisi
avg_GrilledChicken_vita = round(GrilledChicken_Nut['Vit A'].mean(axis=0), 2)
avg_GrilledChicken_vitc = round(GrilledChicken_Nut['Vit C'].mean(axis=0), 2)
avg_GrilledChicken_calc = round(GrilledChicken_Nut['Calcium'].mean(axis=0), 2)
avg_GrilledChicken_iron = round(GrilledChicken_Nut['Iron'].mean(axis=0), 2)

print("Rata-rata VIT A, VIT C, CALCIUM, IRON di GRILLED CHICKEN: ", 
      avg_GrilledChicken_vita, avg_GrilledChicken_vitc, avg_GrilledChicken_calc, avg_GrilledChicken_iron)
print ("Kalori di Grilled Chicken Rata-rata: ",avg_GrilledChicken_Cal)
```

```python
#grafik bar rata2 nutrisi

avg_vita = [avg_CrispyChicken_vita, avg_GrilledChicken_vita]
avg_vitc = [avg_CrispyChicken_vitc, avg_GrilledChicken_vitc]
avg_calc = [avg_CrispyChicken_calc, avg_GrilledChicken_calc]
avg_iron = [avg_CrispyChicken_iron, avg_GrilledChicken_iron]
index = ['Crispy Chicken', 'Grilled Chicken']
avg_vm = pd.DataFrame({'Avg Vit A': avg_vita,
                   'Avg Vit C': avg_vitc,
                   'Avg Calcium': avg_calc,
                   'Avg Iron': avg_iron}, index=index)
ax = avg_vm.plot.bar(rot=0, colormap='cool', figsize=(10,5))
ax.set_title("RATA-RATA NUTRISI DI CRISPY DAN GRILLED CHICKEN")
plt.show()
```

```python
#membuat grafik bar antara Crispy dan Grilled

avg_cal = pd.DataFrame({'Jenis':['Crispy Chicken', 'Grilled Chicken'], 'Rata-rata Kalori': [avg_CrispyChicken_Cal, avg_GrilledChicken_Cal]})
ax = avg_cal.plot.bar(x = 'Jenis', y = 'Rata-rata Kalori')
ax.set_title("Rata-Rata Kalori di Cripy Chicken dan Grilled Chicken")
plt.show()
```

```python
#membuat grafik pie antara cripy dan grilled 

labels = 'Crispy Chicken', 'Grilled Chicken'
sizes = [avg_CrispyChicken_Cal, avg_GrilledChicken_Cal]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Kalori di Cripy Chicken dan Grilled Chicken')
plt.axis('equal')
plt.show()
```

# Soal No 5
Berapa banyak menu yang dapat dipesan untuk memenuhi kebutuhan nutrisi sehari-hari?

```python
#memfilter data berdasarkan nutrisi yang dibutuhkan dalam sehari
nutrisi_sehari = df.iloc[:, [1, 6, 8, 11, 13, 15, 17, 20, 21, 22, 23]]
nonitem = df.iloc[:, [6, 8, 11, 13, 15, 17, 20, 21, 22, 23]]
nutrisi_sehari.head()
```

```python
#data statistik dari nutrisi_sehari
stat_nutrisi = nutrisi_sehari.describe()
stat_nutrisi
```

```python
#memfilter menu yang sesuai dengan kebutuhan nutrisi
index_keb =[]
for i in stat_nutrisi.columns:
    index_keb.extend(nonitem[nonitem[i]==stat_nutrisi.loc['max',i]].index.tolist())
set_index_keb = set(index_keb)
index_keb_new = list(set_index_keb)
```

```python
#menampilkan menu yang memenuhi kebutuhan nutrisi harian
menu_keb_nut = nutrisi_sehari.iloc[index_keb_new, :]
menu_keb_nut
```

```python
#menambahkan kolom total nutrisi harian
nutrisi_data = menu_keb_nut.iloc[:, 2:]
nutrisi_data['Total Daily Nutrition'] = nutrisi_data.sum(axis=1)
nutrisi_data['Kode Menu'] = index_keb_new
nutrisi_data['Item'] = menu_keb_nut['Item']
nutrisi_data
```

```python
#grafik bar plot total nutrisi yang terkandung dalam tiap menu
import seaborn as sns
plt.figure(figsize=(15,10))
sns.barplot(nutrisi_data['Kode Menu'], nutrisi_data['Total Daily Nutrition'], hue=nutrisi_data['Item'])
plt.show()
```

# Soal No 6 (Soal Tambahan)
Berapa kadar gula  yang terkandung dalam setiap menu McD?

```python
#filter data sesuai kandungan sugar
sugar = df.iloc[:, [1, 18]]
sugar.head()
```

```python
#mengelompokkan data berdasarkan kelas nya
class_sugar = []
for i in sugar.Sugars:
    if i < 50:
        class_sugar.append("less sugar")
    elif i == 50:
        class_sugar.append("enough")
    else:
        class_sugar.append("high sugar")
sugar["class_sugar"] = class_sugar
angka = pd.DataFrame(sugar.groupby(["class_sugar"])["class_sugar"].count())
angka.columns = ["count_class_sugar"]
angka
```

```python
sugar.head()
```

```python
#grafik pie menampilkan data gula
fig = plt.figure()
labels = ['Enough', 'High Sugar', 'Less Sugar']
sizes = angka
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Kadar gula dalam menu McD')
plt.show()
```
